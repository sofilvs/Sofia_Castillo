Suponga que usted es un ayudante de un laboratorio de física. En un experimento se pide medir la posición de una masa que cae en un plano inclinado a intervalos regulares en el tiempo. El equipo disponible puede medir las posiciones con una precisión de 4mm. Uno de los estudiantes usa los siguientes datos (en cm):

datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]

Grafique estos datos con sus barras de error. Explique por qué el estudiante merece una nota de 1.0. Compare con los siguientes datos obtenidos por un estudiante que merece 7.0.

datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]

```python
#Importamos las bibliotecas para graficar
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#Escribimos los datos necesarios para crear el gráfico
datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]
datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]

#Tiempo de intervalos regular de tiempo (1 segundo), del mismo tamaño que "datos"
t = np.arange(len(datos))

#Error de +-0.4 cm (debido a la precisión de medición) 
err = 0.4 

# Datos iniciales para comparar con gráfica de plano inclinado "ideal"

g = 9.8  # Aceleración debida a la gravedad
ang = 4.5 # Ángulo de inclinación del plano (en grados)

# Cálculo de la posición ideal en el plano inclinado
h =  0.5 * g * t**2 * np.sin(np.radians(ang))
```

```python
# Grafica posición ideal y posición de los datos nota 1.0

fig = plt.figure()
ax = fig.add_subplot(1,1,1) #tamaño de la figura

ax.scatter(t ,datos, label = 'Estudiante 1') #muestra como puntos cada dato

#Barras de error de los datos
plt.errorbar(t,datos, yerr = err, label = "Error 1", ecolor= 'blue', marker='o') 

#Gráfica ideal (naranja)
plt.plot(t, h, label='Altura (m)', color = "lightblue") 

#Nombra los ejes
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')

#Título y guía del gráfico
plt.legend()
plt.title('Fig 1: Posición en un Plano Inclinado (sin fricción)')
plt.savefig('fig1.jpg') #guarda el gráfico

#Muestra el gráfico
plt.show()
```

```python
#Gráfica comparativa entre el Estudiante 1 (datos) y el Estudiante 2 (datos_7)

#Crea la gráfica
fig= plt.figure()
ax = fig.add_subplot(1,1,1) #tamaño de la figura

#agregar "t" en la posición x y "datos" en la posición y
ax.scatter(t,datos, label = 'Estudiante 1') 
ax.scatter(t,datos_7, label = 'Estudiante 2')

#Nombra los ejes junto con su dimensión
plt.xlabel('$Tiempo (s)$')
plt.ylabel('$Datos (cm) $')

#Barras de error de cada gráfica (junto con colores)
plt.errorbar(t,datos, yerr = err, label = "Error 1", ecolor= 'blue', marker='o')
plt.errorbar(t,datos_7, yerr = err, label= 'Error 2',ecolor='red', marker='o')

#Título, guía del gráfico.
plt.title ('Fig 2: Medición posición de una masa en plano inclinado')
plt.legend()
plt.savefig('fig2.jpg') #guarda el gráfico
plt.show() #muestra el gráfico
```

El estudiante 1 merece un 1.0 no por la mala precisión de datos. Como se puede apreciar en la Fig 1. la curva no parece haber sido creada por mediciones experimentales, sino que parece ser hecha con datos utópicos, es decir, se conocía la gráfica a la que se quería llegar y se rellenaron datos que coincidían con la curva. En la Fig 2 la curva del Estudiante 2 es más errática y menos suave. Los datos se mantienen dentro de los rangos de error y no es una curva perfecta, lo que demuestra que son datos experimentales/reales. En datos experimentales siempre existe un margen de error dentro de los rangos propuestos. 

# Calificacion: 7.0
